# Hype Parameter Tuning

```python
import numpy as np
trees=np.random.randint(20,110,25)
```

```python
trees #trees written instead of n_estimators
```

```python
criterion=['gini','entropy']
max_depth=np.random.randint(1,15,25)#Total depth of tree
min_samples_split=[2,3,4]
max_features=['sqrt','log2']
```

IDcreation

```python

ids={
    'n_estimators':trees,
    'criterion':criterion,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'max_features':max_features,
}
```

```python
print(ids)
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
gcv=GridSearchCV(clf,ids) # n_estimators=clf
```

```python
gcv.fit(xtrain,ytrain)
```

```python
gcv.cv_results_
```

```python
cv_result=pd.DataFrame(gcv.cv_results_)# dataFrame creation by cv_result
```

```python
cv_result
```

```python
#Finding best score
gcv.best_score_
```

```python
gcv.best_params_# finding best parameter
```
